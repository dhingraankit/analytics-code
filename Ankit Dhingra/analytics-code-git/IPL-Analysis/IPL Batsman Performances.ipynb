```python
import pandas as pd
```

```python
balls = pd.read_csv("./train/train_ball_by_ball_data.csv")
balls['player_dismissed_id'] = balls['player_dismissed_id'].fillna(0)
balls['player_dismissed'] = balls['player_dismissed'].fillna("Not Out")
balls['dismissal_kind'] = balls['dismissal_kind'].fillna("Not Out")

#Adding Season# and Match Number to the above dataframe
balls['season'] = balls.match_id.str[:4]
balls['match_number'] = balls.match_id.str[5:]
```

```python
#Total runs by all batsmen:
BatsmanTotalRuns = balls.groupby('batsman')["batsman_runs"].sum().sort_values(ascending=False)
BatsmanTotalRuns = BatsmanTotalRuns.reset_index(name='Runs')

#Total Matches by all batsmen:
BatsmanTotalMatches = balls.groupby('batsman')["match_id"].nunique().sort_values(ascending=False)
BatsmanTotalMatches = BatsmanTotalMatches.reset_index(name='Matches')

#Total Balls by all batsmen:
BatsmanTotalBalls = balls.groupby('batsman')["ball"].count().sort_values(ascending=False)
BatsmanTotalBalls = BatsmanTotalBalls.reset_index(name='BallsFaced')

#Joining the 3 dataframes together:
BatsmanTotalRuns = BatsmanTotalRuns.merge(BatsmanTotalMatches, how ='inner', on="batsman")
BatsmanTotalRuns = BatsmanTotalRuns.merge(BatsmanTotalBalls, how ='inner', on="batsman")

#Adding a column for strike rates of respective players
BatsmanTotalRuns["StrikeRate"] = ((BatsmanTotalRuns["Runs"]/BatsmanTotalRuns["BallsFaced"])*100).round(2)
```

```python
#Highest runs scored by each batsmen:
Highest = balls.groupby(['batsman','match_id'])["batsman_runs"].sum()
Highest = Highest.reset_index(name='Highest')
Highest = Highest.groupby('batsman')['Highest'].max()
Highest = Highest.reset_index(name='HighestScore')

#Merging Highest runs to the existing dataframe, BatsmanTotalRuns:
BatsmanTotalRuns = BatsmanTotalRuns.merge(Highest, how ='inner', on="batsman")
```

```python
#Number of Fours by Every Player
FoursDF = balls[balls.batsman_runs == 4] 
Fours = FoursDF.groupby('batsman')['batsman_runs'].count()
Fours = Fours.reset_index(name='Fours')
BatsmanTotalRuns = BatsmanTotalRuns.merge(Fours, how ='inner', on="batsman")


#Number of Sixes by Every Player
SixesDF = balls[balls.batsman_runs == 6] 
Sixes = SixesDF.groupby('batsman')['batsman_runs'].count()
Sixes = Sixes.reset_index(name='Sixes')
BatsmanTotalRuns = BatsmanTotalRuns.merge(Sixes, how ='inner', on="batsman")
BatsmanTotalRuns
```
